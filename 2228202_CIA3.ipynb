```python
class User:
    def __init__(self, username, password, dob):
        self.username = username
        self.password = password
        self.dob = dob

    def validate_user(self):
        with open("/content/drive/MyDrive/Python/InstagramUser.txt", "r") as file:
            for line in file:
                if line.strip() == self.username + "," + self.password + "," + self.dob:
                    return True
            return False

    def add_user(self):
        with open("/content/drive/MyDrive/Python/InstagramUser.txt", "a") as file:
            file.write(self.username + "," + self.password + "," + self.dob + "\n")

    def update_password(self, old_password, new_password):
        with open("/content/drive/MyDrive/Python/InstagramUser.txt", "r") as file:
            lines = file.readlines()

        with open("/content/drive/MyDrive/Python/InstagramUser.txt", "w") as file:
            updated = False
            for line in lines:
                if line.strip() == self.username + "," + old_password + "," + self.dob:
                    file.write(self.username + "," + new_password + "," + self.dob + "\n")
                    print("Password updated successfully!")
                    updated = True
                else:
                    file.write(line)

            if not updated:
                print("Incorrect username or password.")

    def delete_user(self):
        with open("/content/drive/MyDrive/Python/InstagramUser.txt", "r") as file:
            lines = file.readlines()

        with open("/content/drive/MyDrive/Python/InstagramUser.txt", "w") as file:
            deleted = False
            for line in lines:
                if line.strip() == self.username + "," + self.password + "," + self.dob:
                    print("User deleted successfully!")
                    deleted = True
                else:
                    file.write(line)

            if not deleted:
                print("User not found.")

from datetime import datetime

def calculate_age(dob):
    birth_year = int(dob.split("/")[-1])
    current_year = datetime.now().year
    age = current_year - birth_year
    return age


def check_username(username):
    with open("/content/drive/MyDrive/Python/InstagramUser.txt", "r") as file:
        for line in file:
            line = line.strip()
            if username in line:
                return True
        return False


def login():
    try:
        username = input("Enter username: ")
        password = input("Enter password: ")
        dob = input("Enter the DOB: ")
        user = User(username, password, dob)
        if user.validate_user():
            print("Login Successful")
        else:
            print("Invalid Credential")
    except Exception as e:
        print("An error occurred: ")

class InvalidAgeException(Exception):
  "Raised when the age is less than 18"
  pass

def add_user():
    while True:
        username = input("Enter New Username: ")
        if check_username(username):
            print("Username already exists. Try something different.")
        else:
            password = input("Enter New Password: ")
            dob = input("Enter Date of Birth (DOB) in DD/MM/YYYY format: ")
            age = calculate_age(dob)
            try:
              if age < 18:
                raise InvalidAgeException
              else:
                print("User added successfully!")
            except InvalidAgeException:
              print("You must be 18 or older to use this system.")
          
            user = User(username, password, dob)
            try:
                user.add_user()
            except Exception as e:
                print("An error occurred while adding the user:", e)
        break


def update_password():
    username = input("Enter username: ")

    if not check_username(username):
        print("Username not found.")
        return

    old_password = input("Enter current password: ")
    new_password = input("Enter new password: ")
    dob = input("Enter the DOB: ")

    user = User(username, old_password, dob)
    user.update_password(old_password, new_password)


def delete_user():
    username = input("Enter username to delete: ")
    password = input("Enter password: ")
    dob = input("Enter the DOB: ")

    # check if username exists
    if not check_username(username):
        print("Username not found.")
        return

    user = User(username, password, dob)
    user.delete_user()


def main():
    print("Welcome to Instagram")
    print("1. Login")
    print("2. Add User")
    print("3. Update Password")
    print("4. Delete User")
    print("5. Exit")

    while True:
        choice = int(input("Enter your choice: "))
        if choice == 1:
            login()
        elif choice == 2:
            add_user()
        elif choice == 3:
            update_password()
        elif choice == 4:
            delete_user()
        elif choice == 5:
            break
        else:
            print("Invalid input. Try again.")


if __name__ == "__main__":
    main()
```
